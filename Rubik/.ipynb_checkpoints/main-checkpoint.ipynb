## Initializations

```python
directory = 'results/italy'
os.makedirs(directory, exist_ok=True)
```

```python
# Initialize Environment
max_steps=10
env = RubiksCubeEnv(max_steps)
env.scramble_params = 0 #20  # number of random actions to do for scrambling (called in reset), =0 don't scramble the cube

env.screen_width = 600  # you can reduce the screen size if FPS are too low
env.screen_height = 600

env.cap_fps = 10  # env assumes that you are close to this fps (controls might be weird if it's too far away from it)
env._rotation_step = 5  # higher values make the animation of rotating take fewer frames and vice versa, =90 rotations aren't animated
env._look_point = np.array([0,0,25])
env.action_space.n = 19
```

```python
# Initialize empty lists to store the training data
x_train = []
y_train = []
reward_over_rounds_predicted = []
reward_over_rounds_true = []
```

```python
obs_space_dims = env.observation_space.shape[0]
obs_face_dim = int(env.observation_space.shape[0] / 6)
action_space_dims = env.action_space.n
agent = REINFORCE(obs_space_dims, action_space_dims)
my_reward_model = simple_NN(obs_face_dim, 8)
criterion = torch.nn.BCELoss()
```

```python
exp_name = 'experiment_name_step0'
save_path = directory + '/' + exp_name + '_rewardModel.pth'
torch.save(my_reward_model.state_dict(), save_path)
save_path = directory + '/' + exp_name + '_agentModel'
```

## Observe Optimal Trajectory

```python
obs, info = env.reset()
done = False
truncated = False

trajectory = {}
tt = 0
trajectory[f'state{tt}'] = obs_to_state(obs)

while not truncated:
    env.render()
    obs_tensor = torch.tensor(obs, dtype=torch.float32).unsqueeze(0)
    action = agent.sample_action(obs_tensor)
    #actions = actions + [action]
    obs, reward, done, truncated, info = env.step(action)
    tt+=1
    trajectory[f'state{tt}'] = obs_to_state(obs)
    agent.rewards.append(reward)
    agent.update()
print(done)
print(truncated)

env.close()
```

```python
# Save the trajectory
with open(os.path.join(directory, f'{exp_name}_trajectory.pkl'), 'wb') as f:
    pickle.dump(trajectory, f)
```

## Policy Learning

```python
max_steps=10
env_r = RubiksCubeEnv(max_steps, to_render = False)
env_r.scramble_params = 0  # number of random actions to do for scrambling (called in reset), =0 don't scramble the cube
env_r.action_space.n = 10
```

```python
reward_over_episodes_predicted = []
reward_over_episodes_true = []
total_num_episodes = int(6e3)
```

```python
for episode in range(total_num_episodes):
    obs, info = env_r.reset()
    env_r.close()
    done = False
    truncated = False
    reward_episode_true = 0;
    reward_episode_predicted = 0;
    
    while not truncated: # not done and 
        obs_tensor = torch.tensor(obs, dtype=torch.float32).unsqueeze(0)
        action = agent.sample_action(obs_tensor)
        obs, _, done, truncated, info = env_r.step(action)
        obs_tensor = torch.tensor(obs, dtype=torch.float32)
        
        reward_episode_true += score_front_x(obs)        
        reward_predicted = my_reward_model(obs_tensor[9:18]).detach().numpy()
        reward_episode_predicted += reward_predicted
        
        agent.rewards.append(reward_predicted)
    
    reward_over_episodes_true.append(reward_episode_true)
    reward_over_episodes_predicted.append(reward_episode_predicted)
    agent.update()

    if episode % 1000 == 0:
        avg_reward = np.mean(reward_over_episodes_true[-1000:])
        print("Episode:", episode, "Average True Reward:", avg_reward)
        avg_pred_reward = np.mean(reward_over_episodes_predicted[-1000:])
        print("Episode:", episode, "Average Predicted Reward:", avg_pred_reward)

reward_over_rounds_true = reward_over_rounds_true + reward_over_episodes_true
reward_over_rounds_predicted = reward_over_rounds_predicted + reward_over_episodes_predicted 
```

```python
exp_name = 'experiment_name_step1'
save_path = directory + '/' + exp_name + '_rewardModel.pth'
torch.save(my_reward_model.state_dict(), save_path)
save_path = directory + '/' + exp_name + '_agentModel'
agent.save(save_path)
save_path = directory + '/' + exp_name + '_trueRewards.pth'
torch.save(reward_over_rounds_true, save_path)
save_path = directory + '/' + exp_name + '_predictedRewards.pth'
torch.save(reward_over_rounds_predicted, save_path)
```

```python
window_size = 150
rewards_env = np.array(reward_over_rounds_true).squeeze()
N_episodes = len(rewards_env)
smoothed_episodes = np.arange(window_size // 2 + 1, (N_episodes + 1) - window_size // 2)
smoothed_rewards_env = np.convolve(rewards_env, np.ones(window_size)/window_size, mode='valid')

# Plot the smoothed mean rewards
plt.figure(figsize=(10, 6))
plt.plot(smoothed_episodes, smoothed_rewards_env[:-1], label='')
plt.xlabel('Episodes')
plt.ylabel('True Reward (All Interactions)')
plt.grid(True)
plt.show()
```
